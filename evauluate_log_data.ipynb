```python
def bamp(H):
    """
    A simple converter from H to B, useful to know what the amplitude of the sinc pulse was in Tesla

    Parameters:
    H, 1darray: the field in one direction over time

    Returns:
    int: the maximum of H converted to Tesla
    """
    return int(np.max(H)*sc.mu_0*1000)
```

### FFT PART

```python
step = 1*10**(-11)
print(t[1]-t[0], step)
print("If these values overlap then proceed")
```

```python
def cleanfft0(ar):
    ar = np.abs(sp.fft.rfft(ar))
    ar = np.abs(ar - ar.mean())**2
    return ar
```

```python
def cleanfft1(ar):
    ar = np.abs(sp.fft.rfft(ar))**2
    return ar
```

```python
def cleanfft2(ar):
    ar = np.abs(sp.fft.rfft(ar))
    ar = np.abs(ar - ar.mean())
    return ar
```

```python
def cleanfft3(ar):
    ar = np.abs(sp.fft.rfft(ar))
    return ar
```

```python
#FFT calcs with Sabris suggestions

N = np.size(mx)

f = sp.fft.rfftfreq(N, step)

mfx = cleanfft3(mx)
mfy = cleanfft3(my)
mfz = cleanfft3(mz)
mfx[0:5] = 0
mfy[0:5] = 0
mfz[0:5] = 0
mfs = np.array([mfx, mfy, mfz])
```

```python
ftemp = f[f <= 3e9]
mfztemp = mfz[0: np.size(ftemp)]
indi = np.where(mfztemp == np.max(mfztemp))[0][0]
ftemp[indi]*1e-9
```
